```python
import nltk

from gensim.models import Word2Vec
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
%matplotlib inline
import re
```

```python
text=open("hungergames.txt").read()
```

```python

text = re.sub(r'\[[0-9]*\]',' ',text)
text = re.sub(r'\s+',' ',text)

text = text.lower()

text = re.sub(r'\d',' ',text)
text = re.sub(r'\s+',' ',text)
```

```python
# Preparing the dataset
sentences = nltk.sent_tokenize(text)
sentences
```

```python
sent_word = [nltk.word_tokenize(sentence) for sentence in sentences]
```

```python
sent_word
```

```python


for i in range(len(sent_word)):
    sent_word[i] = [word for word in sent_word[i]
                    if word not in stopwords.words('english')]
```

```python
sent_word
```

```python
# Training the model
model = Word2Vec(sent_word, min_count=1)


words = model.wv.vocab
```

```python
# Test the word vectors
vector = model.wv['younger']

print(vector)
```

```python
# Most similar words
similar = model.wv.most_similar('rotting',topn=5)

similar
```

```python
# You can also check words similarity and compare above
model.wv.similarity(w1='missing',w2='crawling')
```

```python
# filter on non similarity
model.wv.doesnt_match(["crawling","missing","cougar","electrified"])
```

```python
from sklearn.manifold import  TSNE
vocab = ["crawling","missing","electrified","flatten","concealed","district","shackle","wandering","close","open","divide","laugh","arrow","inhaling"]
def tsne_plot(model):
    labels = []
    wordvecs = []

    for word in vocab:
        wordvecs.append(model[word])
        labels.append(word)
    
    tsne_model = TSNE(perplexity=3, n_components=3, init='pca', random_state=42)
    coordinates = tsne_model.fit_transform(wordvecs)

    x = []
    y = []
    for value in coordinates:
        x.append(value[0])
        y.append(value[1])
        
    plt.figure(figsize=(15,8)) 
    for i in range(len(x)):
        plt.scatter(x[i],y[i])
        plt.annotate(labels[i],
                     xy=(x[i], y[i]),
                     xytext=(2, 2),
                     textcoords='offset points', ha='left')

    plt.show()

tsne_plot(model)
```
