```python
#importing libraries
import nltk
import string

from nltk.tokenize import word_tokenize
from collections import Counter

import matplotlib.pyplot as plt
```

```python
#scrapping the tweets before coronovirus was declared as a pandemic
def get_tweets():
    import GetOldTweets3 as got 
    tweetCriteria = got.manager.TweetCriteria().setQuerySearch('coronavirus') \
        .setSince("2020-01-11") \
        .setUntil("2020-02-10") \
        .setMaxTweets(1000)
    # list that contains all tweets containing the word coronavirus
    tweets = got.manager.TweetManager.getTweets(tweetCriteria)
    text_tweets1 = [[tweet.text] for tweet in tweets]
    return text_tweets1
```

```python
# reading text file containing the tweets
text = ""
text_tweets1 = get_tweets()

length = len(text_tweets1)

for i in range(0, length):
    text = text_tweets1[i][0] + " " + text

# converting to lowercase
lower_case1 = text.lower()

```

```python
text_tweets1
```

```python
from nltk.tokenize import regexp_tokenize
regexp_tokenize(lower_case1,"[0-9]+")
```

```python
#the pos_tag defines the parts of speech of each word
pos = nltk.pos_tag(nltk.word_tokenize(lower_case1))
pos
```

```python
#importing stopwords
from nltk.corpus import stopwords

stop_words = set(nltk.corpus.stopwords.words('english'))
len(stop_words)
```

```python
#punctuations are present in the text so importing punctuations to remove the punctuations from the text
string.punctuation
punct =string.punctuation
punct
```

```python
# Empty list to load clean data
cleaned_text1= []

#removing stopwords and punctuations
for word in nltk.word_tokenize(lower_case1):
    if word not in punct:
        if word not in stop_words:
            cleaned_text1.append(word)
    
print ('Original Length of text  == >', len(lower_case1))
print ('length of cleaned text ==>', len(cleaned_text1))
print ('\n',cleaned_text1 )
```

```python
from nltk.stem import WordNetLemmatizer
# Lemmatization for getting root words
lemma_words1 = []
for word in cleaned_text1:
    word = WordNetLemmatizer().lemmatize(word)
    lemma_words1.append(word)
```

```python
# Get emotions text from the directory
emotion_list1 = []
with open('emotions.txt', 'r') as file:
    for line in file:
        clear_line = line.replace('\n', '').replace(',', '').replace("'", '').replace("/", '').replace("*", '').strip()
        word, emotion = clear_line.split(':')
        if word in lemma_words1:
            emotion_list1.append(emotion)
print("People emotions from the text \n", emotion_list1, '\n \n')
w1 = Counter(emotion_list1)
print("Count of each emotion \n", w1)
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
# creating an object of SentimentIntensityAnalyzer 
sia = SentimentIntensityAnalyzer()
```

```python
def sentiment_analyse(sentiment_text):
    score = SentimentIntensityAnalyzer().polarity_scores(sentiment_text)
    if score['neg'] > score['pos']:
        print("\n     ******Negative Sentiment*******")
    elif score['neg'] < score['pos']:
        print("\n     ******Positive Sentiment*******")
    else:
        print("Neutral Sentiment")
```

```python
#BagOfWords Approach
def text_matrix(bog1, countvect):
    terms_doc = countvect.fit_transform(bog1)
    return pd.DataFrame(terms_doc.toarray(),columns=countvect.get_feature_names())
```
